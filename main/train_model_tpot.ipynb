```python
import pandas as pd
```

```python
dataset = pd.read_excel('check.xlsx')
```

```python
dataset
```

```python
from user_transform import UserTransform
```

```python
user_transform = UserTransform()

X = user_transform.transform(dataset)
y = user_transform.target_transform(dataset)
```

```python
X
```

```python
from tpot import TPOTRegressor


tpot = TPOTRegressor(generations=1, population_size=20, verbosity=2, n_jobs = -1, random_state=42)
tpot.fit(X[:40], y[:40])
print('-'*100)
make_pipe, import_libs = tpot.export('', get_pipeline=True)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline

exec(import_libs)
pipe = eval(make_pipe)
pipe = pipe if (type(pipe) == Pipeline) else make_pipeline(pipe)
pipe.steps.insert(0,['usertransform', UserTransform()])

pipe.fit(dataset, user_transform.target_transform(dataset))
pipe.predict(dataset)
pipe.score(dataset, y)
```
